### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# 1. Análisis de facturación

import matplotlib.pyplot as plt

# calcular facturación de cada tienda
facturacion_tienda1 = tienda["Precio"].sum()
facturacion_tienda2 = tienda2["Precio"].sum()
facturacion_tienda3 = tienda3["Precio"].sum()
facturacion_tienda4 = tienda4["Precio"].sum()

# guardar resultados
facturacion = {
    "Tienda 1": facturacion_tienda1,
    "Tienda 2": facturacion_tienda2,
    "Tienda 3": facturacion_tienda3,
    "Tienda 4": facturacion_tienda4
}

# mostrar resultados
print(facturacion)

# gráfico de facturación
plt.bar(facturacion.keys(), facturacion.values())
plt.title("Facturación total por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Facturación")
plt.show()
```

# 2. Ventas por categoría

```python
# contar ventas por categoría en cada tienda

ventas_categoria_tienda1 = tienda["Categoría del Producto"].value_counts()
ventas_categoria_tienda2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria_tienda3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria_tienda4 = tienda4["Categoría del Producto"].value_counts()

print("Ventas por categoría - Tienda 1")
print(ventas_categoria_tienda1)

print("\nVentas por categoría - Tienda 2")
print(ventas_categoria_tienda2)

print("\nVentas por categoría - Tienda 3")
print(ventas_categoria_tienda3)

print("\nVentas por categoría - Tienda 4")
print(ventas_categoria_tienda4)
```

```python
import matplotlib.pyplot as plt

ventas_categoria_tienda1.plot(kind="bar")

plt.title("Ventas por categoría - Tienda 1")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)

plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calificación promedio por tienda

calificacion_tienda1 = tienda["Calificación"].mean()
calificacion_tienda2 = tienda2["Calificación"].mean()
calificacion_tienda3 = tienda3["Calificación"].mean()
calificacion_tienda4 = tienda4["Calificación"].mean()

print("Calificación promedio Tienda 1:", calificacion_tienda1)
print("Calificación promedio Tienda 2:", calificacion_tienda2)
print("Calificación promedio Tienda 3:", calificacion_tienda3)
print("Calificación promedio Tienda 4:", calificacion_tienda4)
```

```python
import matplotlib.pyplot as plt

calificaciones = [
    calificacion_tienda1,
    calificacion_tienda2,
    calificacion_tienda3,
    calificacion_tienda4
]

tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(tiendas, calificaciones)

plt.title("Calificación promedio por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Calificación promedio")

plt.show()
```

# 4. Productos más y menos vendidos

```python
# productos más vendidos por tienda

mas_vendidos_tienda1 = tienda["Producto"].value_counts().head(5)
mas_vendidos_tienda2 = tienda2["Producto"].value_counts().head(5)
mas_vendidos_tienda3 = tienda3["Producto"].value_counts().head(5)
mas_vendidos_tienda4 = tienda4["Producto"].value_counts().head(5)

print("Productos más vendidos - Tienda 1")
print(mas_vendidos_tienda1)

print("\nProductos más vendidos - Tienda 2")
print(mas_vendidos_tienda2)

print("\nProductos más vendidos - Tienda 3")
print(mas_vendidos_tienda3)

print("\nProductos más vendidos - Tienda 4")
print(mas_vendidos_tienda4)
```

```python
# productos menos vendidos por tienda

menos_vendidos_tienda1 = tienda["Producto"].value_counts().tail(5)
menos_vendidos_tienda2 = tienda2["Producto"].value_counts().tail(5)
menos_vendidos_tienda3 = tienda3["Producto"].value_counts().tail(5)
menos_vendidos_tienda4 = tienda4["Producto"].value_counts().tail(5)

print("Productos menos vendidos - Tienda 1")
print(menos_vendidos_tienda1)

print("\nProductos menos vendidos - Tienda 2")
print(menos_vendidos_tienda2)

print("\nProductos menos vendidos - Tienda 3")
print(menos_vendidos_tienda3)

print("\nProductos menos vendidos - Tienda 4")
print(menos_vendidos_tienda4)
```

```python
import matplotlib.pyplot as plt

mas_vendidos_tienda1.plot(kind="bar")

plt.title("Productos más vendidos - Tienda 1")
plt.xlabel("Producto")
plt.ylabel("Cantidad de ventas")

plt.show()
```

# 5. Envío promedio por tienda

```python
# costo promedio de envío por tienda

envio_tienda1 = tienda["Costo de envío"].mean()
envio_tienda2 = tienda2["Costo de envío"].mean()
envio_tienda3 = tienda3["Costo de envío"].mean()
envio_tienda4 = tienda4["Costo de envío"].mean()

print("Costo promedio de envío Tienda 1:", envio_tienda1)
print("Costo promedio de envío Tienda 2:", envio_tienda2)
print("Costo promedio de envío Tienda 3:", envio_tienda3)
print("Costo promedio de envío Tienda 4:", envio_tienda4)
```

```python
import matplotlib.pyplot as plt

envios = [envio_tienda1, envio_tienda2, envio_tienda3, envio_tienda4]
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.bar(tiendas, envios)

plt.title("Costo promedio de envío por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo promedio de envío")

plt.show()
```
